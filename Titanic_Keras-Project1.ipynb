```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
# Loading and reviewing data

df = pd.read_csv("titanic.csv", engine='python')
df.shape
```

```python
# label encoding
from sklearn.preprocessing import LabelEncoder

labelEncoder = LabelEncoder()
df['Sex'] = labelEncoder.fit_transform(df['Sex'].astype(str))

df.head()
```

```python
# Spliting the data
# 100 for testing, rest for training

from sklearn.model_selection import train_test_split

X = df.drop(['Survived', 'Name'], axis=1)
#X = df.drop(['Survived', 'Name', 'Fare'], axis=1)
y = df['Survived']

#split_test_size = 100/887
split_test_size = 100/len(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=split_test_size, random_state=42) 
```

```python
num_inputs = len(X_train.columns.values)
```

```python
X_train = np.asarray(X_train)
X_test = np.asarray(X_test)

y_train = np.asarray(y_train)
y_test = np.asarray(y_test)
```

```python
# Create the keras model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.optimizers import RMSprop
```

```python
# Simple Sequential model
model = Sequential()

# Add a Dense Fully Connected Layer with 1 neuron
#First Hidden Layer
model.add(Dense(512, input_shape=(num_inputs,), activation="relu"))
model.add(Dropout(0.2))
#Second Hidden Layer
model.add(Dense(512, activation="relu"))
model.add(Dropout(0.2))
#Output Layer
#model.add(Dense(1, activation="softmax"))
model.add(Dense(1, activation="sigmoid"))

model.summary()
```

```python
# Compile the model.  Minimize crossentopy for a binary.  Maximize for accuracy
#model.compile(Adam(lr=0.05), 'binary_crossentropy', metrics=['accuracy'])
model.compile(loss='binary_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])
```

```python
#  Fit the model with the data.  Make 200 cycles through the data.

model.fit(X_train, y_train, epochs=100, verbose=1)
```

```python
# Get loss and accuracy on test data
eval_result = model.evaluate(X_test, y_test, verbose=0)
```

```python
# Print test accuracy
print("\n\nTest loss:", eval_result[0], "Test accuracy:", eval_result[1])
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_pred = (y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
